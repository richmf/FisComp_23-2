```python
from pylab import *
```

```python
def primitiva(x,y):
    I_y = zeros(len(y))
    h = x[1] - x[0]
    for i in range(1,len(I_y)-3):
        I_y[i] = 23*y[i-1] +26*y[i] -4*y[i+1] +6*y[i+2] -3*y[i+3] + I_y[i-1]
    I_y[-3] = (23*y[-3] +26*y[-4] -4*y[-5] +6*y[-6] -3*y[-7]) + I_y[-4]
    I_y[-2] = (23*y[-2] +26*y[-3] -4*y[-4] +6*y[-5] -3*y[-6]) + I_y[-3]
    I_y[-1] = (23*y[-1] +26*y[-2] -4*y[-3] +6*y[-4] -3*y[-5]) + I_y[-2]
    return I_y*h/48
```

```python
x = linspace(0,2*pi)
y = sin(x) 
Iy = -cos(x)+1
Iy_num = primitiva(x,y)
plot(x,y,label="$f(x)$")
plot(x,Iy,label="$F(x)$")
plot(x,Iy_num,label="$F(x)$ num")
legend(),grid()
show()
```
