```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,100)
cos_tradicional = np.cos(x)
```

```python
plt.plot(x,cos_tradicional,'--')
plt.show()
```

$$
\cos (x) = \sum _{n=0} ^{\infty} -1^{n}\frac{x^{2n}}{(2n)!} = 1 - \frac{x^2}{2} + \frac{x^4}{24} + ...
$$

$$
\cos (x) = \sum _{n=0} ^{\infty} -1^{n}\frac{x^{2n}}{(2n)!} = 1 - \frac{x^2}{2} + \frac{x^4}{24} + ...
$$

```python
def cos(x,n_max=5):
    y = 0
    for n in range(n_max):
        y += ((-1)**n)*((x**(2*n))/mt.factorial(2*n))
    return y
```

```python
mi_cos = cos(x)
```

```python
plt.plot(x,cos_tradicional,'--')
plt.plot(x,mi_cos,'-')
plt.ylim([-1.1,1.1])
plt.xlim([x.min(),x.max()])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,cos_tradicional,'--',lw=5)
for i in range(1,10):
    plt.plot(x,cos(x,i),label='$N_{max}= %1.0f$' %i)
plt.ylim([-1.1,1.1])
plt.xlim([x.min(),x.max()])
plt.legend(loc='lower left')
plt.show()
```

```python
nombres = ['Rich','Andres','Jean','Regina','Andrea','Raul',np.pi]
for nom in nombres:
    print('Hola %s' %nom)
```

```python
lista_de_numeros = [1,2,3,4,5,3827]
for numerito in lista_de_numeros:
    print('n = %3.1f' %numerito)
```

```python
# paso de listas combinadas
for nom in nombres:
    for numerito in lista_de_numeros:
        print('Hola %s , tu numero es: %1.0f' %(nom,numerito) )
```

```python
# paso de listas combinadas
for (nom,numerito) in zip(nombres,lista_de_numeros):
    print('Hola %s , tu numero es: %1.0f' %(nom,numerito) )
```

```python
nombres_de_archivos = []
for (nom,numerito) in zip(nombres,lista_de_numeros):
    nombres_de_archivos.append(nom+'_'+str(numerito)+'.txt')
print(nombres_de_archivos)
```
