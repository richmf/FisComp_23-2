El sistema de ecuaciones es:
$$
\dot {\vec x } = \vec g(\vec x ; t)
$$

$$
\vec x_{n+1} = \vec x _n + \vec g | _{\vec x_m , t} \delta + \frac{1}{2} \mathbb{J} \cdot \vec g | _{\vec x_m , t} \delta ^2
$$

Probemos con el oscilador armónico
$$
\ddot \theta = - \omega ^2 \theta
$$
con $\theta (0) = \theta _0$ y $\dot \theta (0) = 0$ como condición inicial

$$
\dot \theta = u \\
\dot u = -\omega^2 \theta
$$
Es decir $\vec g(\theta, u) = (u,-\omega^2 \theta)$, con $(\theta_0,0)$ como condición inicial. El Jacobiano del sistema de ecuaciones es:


$$
\mathbb{J}=
\begin{pmatrix}
0 & 1 \\
-\omega ^2 & 0
\end{pmatrix} 
$$

```python
def sist_ec1(x,omega2 = 1):
    # x = theta , u
    return array( [x[1] , -omega2*x[0]] )

def jac_sist_ec1(x,omega2 = 1):
    return array( [ [0,1] , [-omega2,0] ] )
```

```python
def Euler(sistema,ci,t):
    delta = t[1] - t[0]
    resultado = zeros((len(ci),len(t)))
    resultado[:,0] = array(ci) #condicion inicial
    for i in range(len(t)-1):
        # x_n+1 = x_n + g(x_n)*delta
        resultado[:,i+1] = resultado[:,i] + sistema(resultado[:,i])*delta
    return resultado

def Integra_J(sistema,jac,ci,t):
    delta = t[1] - t[0]
    resultado = zeros((len(ci),len(t)))
    resultado[:,0] = array(ci) #condicion inicial
    for i in range(len(t)-1):
        # x_n+1 = x_n + g(x_n)*delta + 0.5*(jac*g)*delta^2
        resultado[:,i+1] = resultado[:,i] + sistema(resultado[:,i])*delta 
        resultado[:,i+1] += (0.5*delta**2)*dot(jac(resultado[:,i]),sistema(resultado[:,i]))
    return resultado
```

```python
t = linspace(0,10*pi,1000)
sol = Euler(sist_ec1,[1,0],t)
sol_j = Integra_J(sist_ec1,jac_sist_ec1,[1,0],t)
plot(t,sol[0,:],label="Euler")
plot(t,sol_j[0,:],'.',label="Jacobiano")
plot(t,cos(t),label="Solucion analítica")
grid,legend()
show()
```

$$
\ddot \theta = -\omega^2 \sin \theta
$$

o bien

$$
\dot \theta = u \\
\dot u = -\omega^2 \sin \theta
$$

$$
\mathbb{J}=
\begin{pmatrix}
0 & 1 \\
-\omega ^2 \cos \theta & 0
\end{pmatrix} 
$$

```python
def pendulo(x,omega2 = 1):
    # x = theta , u
    return array( [ x[1] , -omega2*sin(x[0]) ] )

def jac_pendulo(x,omega2 = 1):
    return array( [ [0,1] , [-omega2*cos(x[0]),0] ] )
```

```python
t = linspace(0,10,1000)
sol_pendulo = Integra_J(pendulo,jac_pendulo,[2,0],t)
plot(t,sol_pendulo[0,:])
show()
```

```python
plot(sol_pendulo[0,:],sol_pendulo[1,:])
show()
```

```python
t = linspace(0,30,2000)
theta0 = linspace(0.1,pi-0.01,13)
for th0 in theta0:
    sol_pendulo = Integra_J(pendulo,jac_pendulo,[th0,0],t)
    plot(sol_pendulo[0,:],sol_pendulo[1,:])
show()
```

$$
\dot P_c(t) = \alpha P_c - \gamma P_c P_z \\
\dot P_z(t) = -\beta P_z + \gamma P_c P_z
$$

```python
def LK(Pob,a=0.5,b=0.3,g=0.4):
    return array( [ a*Pob[0] - g*Pob[0]*Pob[1] , -b*Pob[1] + g*Pob[0]*Pob[1] ] )

def jac_LK(Pob,a=0.5,b=0.3,g=0.4):
    return array([ [a-g*Pob[1] , -g*Pob[0]] , [g*Pob[1] , -b+g*Pob[0] ] ])
```

```python
t = linspace(0,200,10000)
sol = Integra_J(LK,jac_LK,[1,1],t)
plot(sol[0,:],sol[1,:])
show()
```

```python
plot(t,sol[0,:],t,sol[1,:])
show()
```

```python
t = linspace(0,20,1000)
ci = linspace(0.1,1,15)
for Pc0 in ci:
    sol = Integra_J(LK,jac_LK,[Pc0,1],t)
    plot(sol[0,:],sol[1,:])
show()
```
