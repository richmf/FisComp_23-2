```python
from pylab import *
import csv
```

```python
with open('Ejemplo_clase.csv',newline='') as csvfile:
    datos = list(csv.reader(csvfile))
datos = array( datos[1:] , dtype = np.float32)
t , ax , ay , az = datos[:,0]/1e3 , datos[:,1] , datos[:,2] , datos[:,3]
plot(t,ax,label='$a_x$')
plot(t,ay,label='$a_y$')
plot(t,az,label='$a_z$')
legend()
show()
print( sqrt(mean(az)**2 + mean(ay)**2 + mean(ax)**2) )
```
