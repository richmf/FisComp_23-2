$$
a_x(t) = \frac{d v_x}{dt} \implies v_x(t) = v_x(0) + \int _0 ^t a_x(\tilde t) d\tilde t
$$
Voy a usar 
$$
\int_{x_0} ^{x_1} f(x)dx =  \frac{h}{48} ( 23 f_0 + 26f_1 - 4f_2 + 6f_3 - 3f_4)
$$

```python
def int_num_exp(t,dat,ci):
    int_num = zeros(len(t))
    int_num[0] = ci
    for i in range(len(t)-4):
        h_i = (t[i+1] - t[i])/48
        int_num[i+1] = int_num[i] + h_i*(23*dat[i] + 26*dat[i+1] - 4*dat[i+2] + 6*dat[i+3] -3*dat[i+4])
    return int_num
```

```python
vx = int_num_exp(t,ax,0)
vy = int_num_exp(t,ay,0)
plot(t,vy)
```
