```python
from math import *
from pylab import *
%matplotlib inline
```

```python
x = linspace(-2*pi,2*pi,100)
y = tan(x)
y[:-1][diff(y) < 0] = nan
y1 = sin(x)
y2 = cos(x)
```

```python
plot(x,y,label=r'$\tan(x)$')
plot(x,y1,label=r'$\sin(x)$')
plot(x,y2,label=r'$\cos(x)$')
ylim([-2,2]),grid(),legend()
show()
```

```python
def mi_cos(x,N_max=10):
    y = 0
    x = x%(2*pi)
    for n in range(N_max):
        y += ((-1)**n)*((x**(2*n))/factorial(2*n))
    return y
```

```python
%%timeit
cos(pi/4)
```

```python
%%timeit
mi_cos(pi/4,10)
```
