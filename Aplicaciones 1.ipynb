```python
from pylab import *
from scipy.integrate import *
```

```python
def pendulo(t,y,omega=1):
    return array( [ y[1], omega*sin(y[0]) ] )
```

```python
ci = [1,0]
th_dth = solve_ivp(pendulo,(0,10*pi),ci)
th_dth.y[0]
plot(th_dth.y[0],th_dth.y[1])
show()
```

```python
t = linspace(0,10*pi,500)
th_dth = solve_ivp(pendulo,(t[0],t[-1]),ci,dense_output=True)
plot(th_dth.sol(t)[0],th_dth.sol(t)[1])
show()
```
